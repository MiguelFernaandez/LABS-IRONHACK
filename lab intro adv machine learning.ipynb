```python
import pandas as pd
```

```python
data = pd.read_csv("Data/learningSet.csv.txt")
data.shape
```

```python
categorical = data.select_dtypes("object")
```

```python
categorical
```

```python
categorical.drop(columns="OSOURCE", inplace=True)
categorical.drop(columns="ZIP", inplace=True)
categorical.head()
```

```python
nulls = pd.DataFrame(data.isna().sum()/len(data))
nulls= nulls.reset_index()
nulls.columns = ['column_name', 'Percentage Null Values']
nulls.sort_values(by='Percentage Null Values', ascending = False)
```

```python
threshold = 0.85
drop_list = []

for column in categorical.columns:
    null_perc = categorical[column].isna().sum() / len(categorical[column])
    if null_perc > threshold:
        drop_list.append(column)

print(drop_list)
```

```python
categorical.drop(columns=drop_list, inplace=True)
categorical.head()
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
print(categorical['GENDER'].value_counts())
```

```python
def replace_values(value):
    if value not in ["M", "F"]:
        return "others"
    return value

for column in categorical:
    categorical[column] = categorical[column].apply(replace_values)

categorical["GENDER"].value_counts()
```
